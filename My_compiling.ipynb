```python
import os, glob
import pandas as pd
```

```python
path = "resources/"

all_files = glob.glob(os.path.join(path, "JC*.csv"))
```

```python
file_list = []

for f in all_files:
    f_data = pd.read_csv(f, sep=',')
    file_list.append(f_data)
```

```python
file_list
```

```python
df_merged = pd.concat(file_list, ignore_index=True, join ='inner')
df_merged
```

```python
df_merged.to_csv( "merged.csv", index = True)
```

```python
# df_merged.rename(columns={ df_merged.columns[0]: "unique_id" }, inplace = False)
df_merged['unique_id'] = range(1, len(df_merged.index)+1)
df_merged
```

```python
list(df_merged.columns)
```

```python
a_to_b_df = df_merged[['unique_id','start station name','start station latitude','start station longitude','end station name','end station latitude','end station longitude']]
```

```python
a_to_b_df.to_csv( "a_to_b.csv", index =True)
```
